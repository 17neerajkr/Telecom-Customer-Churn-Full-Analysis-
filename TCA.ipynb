```python

import matplotlib.pyplot as plt
import pandas as pd
# Create a cross-tab (counts)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to % of total per SeniorCitizen group
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot
ax = ct_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(5,4),
    edgecolor='black'
)

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.title("Churn % by SeniorCitizen (Stacked)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
# Your categorical columns
cat_cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

# Number of plots
n = len(cat_cols)

# Define grid (3 columns)
ncols = 3
nrows = int(np.ceil(n / ncols))

plt.figure(figsize=(18, nrows * 4))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(nrows, ncols, i)
    sns.countplot(x=col, data=df, hue="Churn")
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```

# 📊 Summary of Categorical Data .

1. InternetService: Fiber optic users show the highest churn. DSL users churn less. No-internet users rarely churn.

2. OnlineSecurity / OnlineBackup / TechSupport / DeviceProtection: Customers with No services churn more. Add-on services reduce churn.

3. Contract: Month-to-Month customers churn the most. One-year and Two-year contracts have very low churn.

4. PaymentMethod: Electronic Check users have high churn. Other payment methods show lower churn.

5. PaperlessBilling: Paperless users churn more.

6. PhoneService / MultipleLines / StreamingTV / StreamingMovies: Small effect on churn compared to other features.
